## Data Cleaning

Removing unnecessary columns that do not contribute to prediction.

```python
df.drop(['id', 'Unnamed: 32'], axis=1, inplace=True)
```

## Label Encoding

Converting categorical diagnosis values into numerical format:
- Malignant = 1
- Benign = 0

```python
df['diagnosis'] = df['diagnosis'].map({'M':1, 'B':0})
```

```python
# Step : define target(y) and features(X)

y = df['diagnosis']

X = df.drop('diagnosis', axis=1)
```

```python
X.shape
```

```python
y.shape
```

## Train-Test Split

The dataset is divided into training and testing sets for model evaluation.

```python
# Step : split data

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.3,
    random_state=2529
)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

## Feature Scaling

Feature scaling standardizes numerical values for improved model performance.

```python
# Step : standardize X variables

from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

X_train = ss.fit_transform(X_train)

X_test = ss.transform(X_test)
```

## Logistic Regression Model

Logistic Regression is used for binary classification in this project.

```python
# Step : select model

from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
```

```python
# Step : train model

model.fit(X_train, y_train)
```

```python
# Step : predict

y_pred = model.predict(X_test)
```

```python
y_pred
```

## Model Evaluation

The trained model is evaluated using accuracy score, classification report, and confusion matrix.

```python
# Step : accuracy score

from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```

```python
# Step : classification report

from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
# Step : confusion matrix

from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True, fmt='d')

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")

plt.show()
```
